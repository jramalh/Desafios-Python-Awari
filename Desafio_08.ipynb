### Desafio 8

Escreva um script Python para encontrar as 10 palavras mais longas em um arquivo de texto.
O arquivo .txt está localizado na mesma pasta do projeto (**texto.txt**).

# Método 1
Da biblioteca string, buscamos o que é pontuacao

```python
# Seu código
import string
pontuacao = string.punctuation
```

Lemos o texto

```python
with open("texto.txt", "r") as f:
    texto = f.read()
```

Substituimos o caracter de quebra de linha (`\n`) por espaco

```python
texto = texto.replace("\n"," ")
```

Usamos list comprehension para remover a pontuacao:
- Para cada letra do texto:
    - Se a letra nao estiver na pontuacao, mantemos
    - Se estiver, substituimos por espaco `" "``

```python
texto_sem_pontuacao = [letra if letra not in pontuacao else " " for letra in texto]
```

O comando anterior gera uma lista, que precisamos juntar com um join

```python
texto_sem_pontuacao = "".join(texto_sem_pontuacao)
```

Buscamos uma lista das palavras únicas usando o set

```python
palavras_unicas = list(set(texto_sem_pontuacao.split(" ")))
```

O comando sorted aceita como key uma funcao, no caso vamos usar `len` (length, tamanho). Ele dá o resultado de maneira crescente. Como queremos as maiores, pegamos então as 10 últimas.

```python
sorted(palavras_unicas, key=len)[-10:]
```
